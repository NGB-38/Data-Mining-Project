```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv("/content/Titanic-Dataset.csv")
df=pd.DataFrame(data)
df
```

```python
# Show the descriptive statistics of all numeric values
analyzed_numeric_attributes= df.select_dtypes(include=['number']).describe()
analyzed_numeric_attributes
```

```python
#Filling missing values with its median
df['Age'] = df['Age'].fillna(df['Age'].median())
df
```

```python
#Show descriptive statistics after filling all NA
analyzed_numeric_attributes2= df.select_dtypes(include=['number']).describe()
analyzed_numeric_attributes2
```

```python
df['Relatives'] = df['SibSp'] +  df['Parch']
```

```python
df.describe()
```

```python
#Make a new dataset after removing unnecessary features
columns_to_exclude = ['PassengerId','Fare','Name','Ticket','Cabin','Embarked','SibSp','Parch']
new_df = df.drop(columns=columns_to_exclude)
new_df
```

```python
sns.boxplot(new_df["Age"],orient='h')
plt.show()
```

```python
sns.boxplot(new_df["Relatives"],orient='h')
plt.show()
```

```python
sns.boxplot(new_df["Pclass"],orient='h')
plt.show()
```

```python
# Visualizing survival count
sns.countplot(x='Survived',data=new_df)
plt.title('Survival Count')
plt.show()
```

```python
# Class distribution (1 :survived and 0: not survived)
sns.countplot(x='Pclass', data=new_df, hue='Survived')
plt.title('Pclass and Survival')
plt.show()
```

```python
# Survival by gender
sns.countplot(x='Sex', hue='Survived', data=new_df)
plt.title('Sex and Survival')
plt.show()
```

```python
# Survived by age
bins = np.arange(0,81, 10)
labels = [f'{i}-{i+10}' for i in bins[:-1]]

new_df['Age_Binned'] = pd.cut(new_df['Age'], bins=bins, labels=labels, right=False)

sns.countplot(x='Age_Binned', hue='Survived', data=new_df)
plt.title('Age and Survival ')
plt.xticks(rotation=45)
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of Age after filling NA
sns.histplot(new_df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
selected_columns=df[['Survived','Pclass','Age','Relatives']]
corr = selected_columns.corr()

plt.figure(figsize=(20, 12), dpi=300)  # Adjust the width and height as per your preference
sns.heatmap(corr, annot=True, fmt='.2f')
plt.show()
```
